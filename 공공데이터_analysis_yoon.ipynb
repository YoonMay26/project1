## 1. 이직 여부에 따른 취업 특성 분석

```python
import pandas as pd

# '이직 여부' 컬럼을 기반으로 'is_job_change' 컬럼 생성 ('1'은 비이직자, '2'는 이직자)
df1['is_job_change'] = df1['이직여부'].apply(lambda x: 1 if x == 2 else 0)

# 그룹별 이직률 계산 함수
def calculate_job_change_rate(df1, group_col):
    grouped_data = df1.groupby(group_col)
    grouped_stats = grouped_data['is_job_change'].agg(['sum', 'count']).reset_index()
    grouped_stats['job_change_rate'] = grouped_stats['sum'] / grouped_stats['count']
    return grouped_stats[[group_col, 'job_change_rate']]

# 각 변수별 이직률 데이터프레임 생성
majorcat_df1 = calculate_job_change_rate(df1, 'majorcat')  # 전공 계열별 이직률
school_df1 = calculate_job_change_rate(df1, 'school')      # 학교 유형별 이직률
area_df1 = calculate_job_change_rate(df1, 'area')          # 지역별 이직률
sex_df1 = calculate_job_change_rate(df1, 'sex')            # 성별별 이직률
```

### 전공계열, 학교유형, 성별에 따른 이직률

```python
import pandas as pd

# '이직 여부' 컬럼을 기반으로 'is_job_change' 컬럼 생성 ('1'은 비이직자, '2'는 이직자)
df1['is_job_change'] = df1['이직여부'].apply(lambda x: 1 if x == 2 else 0)

# majorcat, school, sex를 기준으로 그룹화하여 이직률 계산
grouped_data = df1.groupby(['majorcat', 'school', 'sex'])
grouped_stats = grouped_data['is_job_change'].agg(['sum', 'count']).reset_index()

# 이직률 계산
grouped_stats['job_change_rate'] = grouped_stats['sum'] / grouped_stats['count']

majorcat_school_sex_df1 = grouped_stats[['majorcat', 'school', 'sex', 'job_change_rate']]
majorcat_school_sex_df1
```

### 전공계열과 성별별 이직률

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm 

font_path = 'C:/Windows/Fonts/malgun.ttf'  
font_prop = fm.FontProperties(fname=font_path, size=14)
plt.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False  

plt.figure(figsize=(12, 6))
sns.barplot(data=majorcat_school_sex_df1, x='majorcat', y='job_change_rate', hue='sex')
plt.title('전공 계열과 성별별 이직률')
plt.show()
```

이직률은 교육 전공 여성이 가장 높고 인문 전공 여성이 가장 낮다.

### 학교 유형과 성별에 따른 이직률

```python
plt.figure(figsize=(10, 6))
heatmap_data = majorcat_school_sex_df1.pivot_table(index='school', columns='sex', values='job_change_rate')
sns.heatmap(heatmap_data, annot=True, fmt='.2f', cmap='Blues', cbar_kws={'label': '이직률'})
plt.title('학교 유형과 성별에 따른 이직률 히트맵')
plt.xlabel('성별')
plt.ylabel('학교 유형')
plt.show()
```

이직률은 교육대를 졸업한 여성이 가장 높고 4년제 졸업 여성이 가장 낮다.

```python
# 학교 유형과 전공 계열별 이직률 바 차트

school_majorcat_df1 = grouped_stats[['school', 'majorcat', 'job_change_rate']]
plt.figure(figsize=(12, 6))
sns.barplot(data=school_majorcat_df1, x='school', y='job_change_rate', hue='majorcat')

plt.title('학교 유형과 전공 계열별 이직률')
plt.xlabel('학교 유형')
plt.ylabel('이직률')
plt.show()
```

2~3년제에서는 예체능 전공이 가장 이직률이 낮고 자연 전공이 가장 이직률이 높다.   
4년제에서는 인문 전공이 가장 이직률이 낮고 의약 전공이 가장 이직률이 높다.   
교육대는 데이터가 불충분하므로 비교하기 어렵다.

## 취업 현황 분석

```python
# 전공별 산업군 분포 계산

industry_distribution = df1.pivot_table(index='majorcat', columns='industry', values='job', aggfunc='count', fill_value=0)
industry_distribution_percentage = industry_distribution.div(industry_distribution.sum(axis=1), axis=0) * 100  

plt.figure(figsize=(14, 8))
industry_distribution_percentage.plot(kind='bar', stacked=True)
plt.title('전공 계열별 산업군 분포')
plt.xlabel('전공 계열')
plt.ylabel('진출 비율 (%)')
plt.legend(title='산업군', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#전공별 직무 유형 분포 계산
job_distribution = df1.pivot_table(index='majorcat', columns='job', values='industry', aggfunc='count', fill_value=0)
job_distribution_percentage = job_distribution.div(job_distribution.sum(axis=1), axis=0) * 100  # 비율로 변환

plt.figure(figsize=(14, 8))
job_distribution_percentage.plot(kind='bar', stacked=True)
plt.title('전공 계열별 직무 유형(job) 분포')
plt.xlabel('전공 계열')
plt.ylabel('진출 비율 (%)')
plt.legend(title='직무 유형', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## 만족도 통계

```python

satis_df1 = ['a126', 'a127', 'a128', 'a129', 'a130', 'a131', 'a132', 'a133', 'a134', 'a135', 'a136', 'a137', 'a138', 'a139', 'a140', 'a141']

# 전공 계열별 평균 만족도 계산
major_satis = df1.groupby('majorcat')[satis_df1].mean().reset_index()

# 산업군별 평균 만족도 계산
industry_satis = df1.groupby('industry')[satis_df1].mean().reset_index()

# 직무 유형별 평균 만족도 계산
job_satis = df1.groupby('job')[satis_df1].mean().reset_index()
```

```python
job_satis
```

```python

import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
bar_width = 0.075  
x = np.arange(len(major_satis['majorcat'])) * 1.5  


for i, column in enumerate(satis_df1):
    plt.bar(x + i * bar_width, major_satis[column], width=bar_width, label=column)

plt.title('전공 계열별 평균 만족도')
plt.xlabel('전공 계열')
plt.ylabel('평균 만족도')
plt.xticks(x + bar_width * (len(satis_df1) - 1) / 2, major_satis['majorcat'], rotation=45, ha='right')  # x축 레이블 설정
plt.legend(title='만족도 항목', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

전공 계열 중 '교육'전공 계열이 평균적으로 가장 높은 만족도를 보였다.

### 산업군별 만족도 바 차트

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


num_columns = len(industry_satis.columns) - 1  
colors = plt.cm.viridis(np.linspace(0, 1, num_columns))  

# 산업군별 평균 만족도를 바 차트로 시각화
plt.figure(figsize=(14, 8))

bar_width = 0.05 
x = np.arange(len(industry_satis['industry']))  

for i, column in enumerate(industry_satis.columns[1:]):  
    plt.bar(x + i * bar_width, industry_satis[column], 
            width=bar_width, label=column, color=colors[i])

plt.title('산업군별 평균 만족도')
plt.xlabel('산업군')
plt.ylabel('평균 만족도')
plt.xticks(x + bar_width * (num_columns - 1) / 2, 
           industry_satis['industry'], rotation=45) 
plt.legend(title='만족도 항목', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
