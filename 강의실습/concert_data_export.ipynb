```python
import requests
from bs4 import BeautifulSoup


url = 'https://tickets.interpark.com/contents/genre/concert'
response = requests.get(url)


soup = BeautifulSoup(response.text, 'html.parser')


css_selector = "#contents > article.ProductList_productListWrapper__frZE_.GProductList.Panel_wrap__b444b > section > div.Panel_contents__f7025.Panel_fill-scroll__8271f > div.ProductList_contents__eUxgq.InfiniteList_list__3c511.InfiniteList_column-desktop-5__9cab3.InfiniteList_column-mobile-1__853f7.InfiniteList_ticket-list__dfe68"
div_element = soup.select_one(css_selector)

if div_element:
    print(div_element.prettify())
else:
    print("요소를 찾을 수 없습니다.")
```

```python
from bs4 import BeautifulSoup
import re



concerts = soup.find_all('a', class_='TicketItem_ticketItem__')

print("| 공연명 | 장소 | 날짜 |")
print("|--------|------|------|")

for concert in concerts:
    name = concert.find('li', class_='TicketItem_goodsName__Ju76j').text.strip()
    place = concert.find('li', class_='TicketItem_placeName__ls_9C').text.strip()
    date = concert.find('li', class_='TicketItem_playDate__5ePr2').text.strip()
    
    
    if len(name) > 30:
        name = name[:27] + "..."
    
    print(f"| {name} | {place} | {date} |")
```

```python
from bs4 import BeautifulSoup
import csv


concerts = soup.find_all('a', class_='TicketItem_ticketItem__')


with open('concerts.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['공연명', '장소', '날짜'])  

    for concert in concerts:
        name = concert.find('li', class_='TicketItem_goodsName__Ju76j').text.strip()
        place = concert.find('li', class_='TicketItem_placeName__ls_9C').text.strip()
        date = concert.find('li', class_='TicketItem_playDate__5ePr2').text.strip()
        
        writer.writerow([name, place, date])

print("concerts.csv 파일이 생성되었습니다.")
```

```python
import pandas as pd
import mysql.connector


csv_file = 'concerts.csv'
df = pd.read_csv(csv_file)


connection = mysql.connector.connect(
    host='localhost',  
    user='root',  
    password='admin',  
    database='my_db'  
)

cursor = connection.cursor()


create_table_query = """
CREATE TABLE IF NOT EXISTS concerts (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(255),
    place VARCHAR(255),
    date VARCHAR(255)
)
"""
cursor.execute(create_table_query)


for index, row in df.iterrows():
    insert_query = """
    INSERT INTO concerts (name, place, date) VALUES (%s, %s, %s)
    """
    cursor.execute(insert_query, (row['공연명'], row['장소'], row['날짜']))


connection.commit()


cursor.close()
connection.close()

print("CSV 파일의 데이터가 MySQL 데이터베이스에 성공적으로 삽입되었습니다.")
```
