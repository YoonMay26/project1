```python
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
print("원본 데이터프레임:")
print(df)

# 나이에 따라 직원 분류
def classify_age(df):
    young = []
    mid = []
    experienced = []

    for index, row in df.iterrows():
        if row['Age'] < 30:
            young.append(row['Name'])
        elif row['Age'] < 40:
            mid.append(row['Name'])
        else:
            experienced.append(row['Name'])

    return young, mid, experienced

young, mid, experienced = classify_age(df)

print("\n연령대별 직원 분류:")
print("젊은 직원:", young)
print("중간 직원:", mid)
print("경험 많은 직원:", experienced)
```

```python
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
print("원본 데이터프레임:")
print(df)

# 나이에 따라 직원 분류
def classify_age(df):
    young = []
    mid = []
    experienced = []

    for index, row in df.iterrows():
        if row['Age'] < 30:
            young.append(row['Name'])
        elif row['Age'] < 40:
            mid.append(row['Name'])
        else:
            experienced.append(row['Name'])

    return young, mid, experienced

young, mid, experienced = classify_age(df)

# 결과를 데이터프레임으로 변환
classification_data = {
    'Age Group': ['Young', 'Mid', 'Experienced'],
    'Employees': [young, mid, experienced]
}

classification_df = pd.DataFrame(classification_data)

print("\n연령대별 직원 분류 데이터프레임:")
print(classification_df)
```

```python
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
print("원본 데이터프레임:")
print(df)

# 나이에 따라 직원 분류
def classify_age(age):
    if age < 30:
        return 'Young'
    elif age < 40:
        return 'Mid'
    else:
        return 'Experienced'

# 'Age Group' 컬럼 추가
df['Age Group'] = df['Age'].apply(classify_age)

print("\n연령대가 추가된 데이터프레임:")
print(df)
```
